```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('C:/Users/Specter/Desktop/resources/Data Files/3. ST Academy - Decision Trees resource files/Movie_classification.csv', header = 0)
df['Time_taken'].mean()
df['Time_taken'].fillna(value = df['Time_taken'].mean(), inplace= True  )
df.info()
df.head()
df = pd.get_dummies(df, columns= ['3D_available','Genre'], drop_first= True)
X = df.loc[:, df.columns != 'Start_Tech_Oscar']
y = df['Start_Tech_Oscar']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
X_train.shape
from sklearn import tree
clftree = tree.DecisionTreeClassifier(max_depth = 3)
clftree.fit(X_train,y_train)
y_train_pred = clftree.predict(X_train)
y_test_pred = clftree.predict(X_test)
y_test_pred
from sklearn.metrics import accuracy_score, confusion_matrix
confusion_matrix(y_train, y_train_pred)
confusion_matrix(y_test, y_test_pred)
accuracy_score(y_test, y_test_pred)
#Plotting Decision Tree is Very Difficult in Python
#Below is the code mentioned for it
#If using anaconda.
#Open anaconda prompt, write conda install pydotplus.
#If not using anaconda,
#Open Command prompt and write pip install pydotplus
dot_data = tree.export_graphviz(clftree,out_file=None, feature_names= X_train.columns, filled= True)
from IPython.display import Image
import pydotplus
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```
