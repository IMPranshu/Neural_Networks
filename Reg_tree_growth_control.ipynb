```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('C:/Users/Specter/Desktop/resources/Data Files/3. ST Academy - Decision Trees resource files/Movie_regression.csv', header = 0)
df['Time_taken'].mean()
df['Time_taken'].fillna(value = df['Time_taken'].mean(), inplace= True  )
df.info()
df.head()
df = pd.get_dummies(df, columns= ['3D_available','Genre'], drop_first= True)
X = df.loc[:, df.columns != 'Collection']
y = df['Collection']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
X_train.shape
from sklearn import tree
regtree = tree.DecisionTreeRegressor(max_depth=3)
regtree.fit(X_train,y_train)
y_train_pred = regtree.predict(X_train)
y_test_pred = regtree.predict(X_test)
#Plotting Decision Tree is Very Difficult in Python
#Below is the code mentioned for it
#If using anaconda.
#Open anaconda prompt, write conda install pydotplus.
#If not using anaconda,
#Open Command prompt and write pip install pydotplus
dot_data = tree.export_graphviz(regtree, out_file=None)
from IPython.display import Image
import pydotplus
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

```python
regtree1 = tree.DecisionTreeRegressor(max_depth=3)
regtree1.fit(X_train, y_train)
dot_data = tree.export_graphviz(regtree1,out_file=None, feature_names= X_train.columns, filled= True)
graph1 = pydotplus.graph_from_dot_data(dot_data)
Image(graph1.create_png())
```

```python
regtree1 = tree.DecisionTreeRegressor(min_samples_split= 40)
regtree1.fit(X_train, y_train)
dot_data = tree.export_graphviz(regtree1,out_file=None, feature_names= X_train.columns, filled= True)
graph1 = pydotplus.graph_from_dot_data(dot_data)
Image(graph1.create_png())
```
